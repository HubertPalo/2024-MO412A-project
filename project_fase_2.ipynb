```python
import requests
from bs4 import BeautifulSoup
import datetime
import pandas as pd
from tqdm import tqdm
import os
```

```python
os.makedirs('files', exist_ok=True)
data = pd.read_csv('document_links_from_2020-01-01_to_2024-05-01.csv')
data
```

```python
for key, row in data.iterrows():
    print(key, row['document'], row['link'])
    try:
        # Request page
        page = requests.get(row['link'] + '/votacoes')
        # Create a BeautifulSoup object
        soup = BeautifulSoup(page.content, 'html.parser')
        # Check if the page contains a 'a href' with the text 'Por nome'
        file_url = soup.find('a', string='Por nome').attrs['href']
        print(file_url)
        # Request the document page
        request_result = requests.get(file_url)
        file_name = row["document"].replace('/', '_') + '.pdf'
        print(f'Saving {file_name}')
        with open(f'files/{row["document"].replace("/", "_")}.pdf', 'wb') as f:
            f.write(request_result.content)
    except Exception as e:
        # Print the error
        print('Error', e)
```
