```python
import requests
from bs4 import BeautifulSoup
import datetime
import pandas as pd
from tqdm import tqdm
import os
```

```python
os.makedirs('files', exist_ok=True)
data = pd.read_csv('document_links_from_2020-01-01_to_2024-05-01.csv')
data
```

```python
from tqdm import tqdm

rows = []
for key, row in tqdm(data.iterrows(), total=len(data)):
    page = requests.get(row['link'] + '/votacoes')
    # Create a BeautifulSoup object
    soup = BeautifulSoup(page.content, 'html.parser')
    # Check if the page contains a 'a href' with the text 'Por nome'
    file_urls = soup.find_all('a', string='Por nome')
    rows.append({
        'document': row['document'],
        'link': row['link'],
        'file_urls': len(file_urls) > 0
    })
    filtered_data = pd.DataFrame(rows)
    filtered_data = filtered_data[filtered_data['file_urls'] == True]
    filtered_data.to_csv('document_links_filtered.csv', index=False)
```
